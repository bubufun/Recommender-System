## Data for Customized Recommendation-BASKET_SUM_COMM

```python
import pandas as pd
import os
df = pd.read_csv(r'./transaction_data.csv')
hid_list = list(set(df['household_key'].tolist()))
```

```python
df5 = pd.read_csv(r'./product.csv')
dfs = {}
dfs["transaction_data_2"]=pd.merge(df,df5, on='PRODUCT_ID').sort_values(by=['BASKET_ID'])
df6 = dfs["transaction_data_2"]
```

```python
if not os.path.exists(r'./data_processed-cid_SUB_COMM/'):
    os.mkdir(r'./data_processed-cid_SUB_COMM/')
```

```python
for i in hid_list:
    temp_list = list(df6[df6['household_key'] == i].groupby('BASKET_ID')['SUB_COMMODITY_DESC']) 
    BID_list = []
    SUB_COM_list = []
    for BD, SUB in temp_list:
        BID_list.append(BD)
        SUB_COM_list.append(list(SUB))
        data_dict =  {'BASKET_ID': pd.Series(BID_list),'words': pd.Series(SUB_COM_list)}
        temp_df = pd.DataFrame(data_dict)
        temp_df.to_csv('./data_processed-cid_SUB_COMM/SUB_COMM_CID_%s.csv' % i ,index =False)
```
